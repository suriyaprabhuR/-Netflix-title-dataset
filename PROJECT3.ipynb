##Decting how many no of outliers


 IQR (interqurantile range) is used to detect the outliers in all numerical columns.

Steps :

* calculate the Q1(25th percentile) and Q3(75th percentile).

find IQR=Q3-Q1

* define lower bound and upper bound

lower bound : q1-1.5*IQR

upper bound : q3+1.5*IQR

any value outside this range is considered as outliers

```python
net.describe() # to describe the numerical column for find outliers
```

# observation
  
  It shows the numerical column at (released_year) column to find the outliers.

```python
# to find Q1 and Q3 values

Q1 = net['release_year'].quantile(0.25)
Q3 = net['release_year'].quantile(0.75)
display(Q1)
display(Q3)
```

```python
# to find IQR

IQR = Q3-Q1 # it will be synatx for find IQR
display(IQR)
```

```python
# to find lower and upper limit to set boundries

lower_limit = Q1-1.5*IQR
upper_limit = Q3+1.5*IQR
display(lower_limit)
display(upper_limit)
```

```python
# to find outliers

outliers = net[(net['release_year']<lower_limit) | (net['release_year']>upper_limit)]
display(outliers)
```

# observation

  there are 473 rows are outliers in the release_year column.

```python
# to drop the outliers
net.drop(outliers.index,inplace=True)
```

```python
# to reset the index value
net.reset_index(drop=True,inplace=True)
```

```python
#verify the outlier drop or not
net.shape[0]
```

# observation
  
  After drop the outliers there are 4859 rows in the dataset.
